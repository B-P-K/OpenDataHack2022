```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import geopy.distance
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import NearestNeighbors
```

# INFO Adds column for safety score of nearest street to data/gwr_cleaned_wgs84.csv

```python
df_street_security = pd.read_csv("data/wahrgenommene-sicherheit-stadt-stgallen-streetwise.csv", delimiter=';')
df_geo_admin = pd.read_csv("data/merged.csv")
```

```python
df_geo_admin
```

```python
geopy.distance.geodesic((47, 30), (48, 30)).m
```

```python
neigh = NearestNeighbors(n_neighbors=1, metric=lambda a, b: geopy.distance.geodesic(a, b).m)
neigh.fit(df_street_security[['lat', 'lng']].to_numpy())
```

```python
idx = neigh.kneighbors(np.stack([df_geo_admin['latitude'], df_geo_admin['longitude']], axis=-1), return_distance=False)
```

```python
df_street_security[df_street_security.index.duplicated()]
```

```python
df_geo_admin = df_geo_admin.reset_index()
```

```python
df_geo_admin
```

```python
df_geo_admin['street_safety_score'] = list(df_street_security.iloc[idx.flatten()]['score'])
```

```python
df_geo_admin.to_csv("data/merged.csv")
```
